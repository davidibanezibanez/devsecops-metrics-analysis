# Convert .bib to .csv for ACM and IEEE

```python
import bibtexparser
import pandas as pd

# Read BibTeX file
with open('data-science-searches/links/links_ieee.bib') as bibfile:
    bib_database = bibtexparser.load(bibfile)

# Convert to DataFrame and save as CSV
df = pd.DataFrame(bib_database.entries)
df.to_csv('data-science-searches/links/links_ieee.csv', index=False)
```

# Remove unnecessary columns while keeping "title", "authors", "doi", "source", "year", and "keywords"

```python
import os
import pandas as pd

# Directories
input_dir = "data-science-searches/linksnormalizedheaders/"
output_dir = "data-science-searches/linksnormalizedheadersclean/"
os.makedirs(output_dir, exist_ok=True)

# Files to process
files = [
    "links_acm_normalized.csv",
    "links_ieee_normalized.csv",
    "links_scopus_normalized.csv",
    "links_springerlink_normalized.csv"
]

# Columns to keep
columns_to_keep = ["title", "authors", "doi", "source", "year", "keywords"]

# Process each file
for file in files:
    input_path = os.path.join(input_dir, file)
    df = pd.read_csv(input_path)

    # Keep only required columns (ignore missing ones)
    df_clean = df[[col for col in columns_to_keep if col in df.columns]]

    # Output file name
    output_file = file.replace("_normalized.csv", "_normalized_clean.csv")
    output_path = os.path.join(output_dir, output_file)

    # Save cleaned CSV
    df_clean.to_csv(output_path, index=False, encoding="utf-8")

    print(f"Processed and saved file at: {output_path}")
```

# Add "database_source" column

```python
import pandas as pd
import os

# Configuration
nombre_archivo_original = "links_springerlink_normalized_clean.csv"  # Adjust for each CSV
valor_database_source = "springerlink"  # Adjust for each CSV

# Input path
ruta_entrada = os.path.join("data-science-searches/linksnormalizedheadersclean", nombre_archivo_original)

# Load CSV
df = pd.read_csv(ruta_entrada)

# Add new column
df["database_source"] = valor_database_source

# Output file name
nombre_base = os.path.splitext(nombre_archivo_original)[0]
nuevo_nombre_archivo = f"{nombre_base}_withdatabasesource.csv"

# Output directory
carpeta_salida = "data-science-searches/linksnormalizedheaderscleanwithdatabasesource"
os.makedirs(carpeta_salida, exist_ok=True)

# Output path
ruta_salida = os.path.join(carpeta_salida, nuevo_nombre_archivo)

# Save CSV
df.to_csv(ruta_salida, index=False)

print(f"New CSV saved at: {ruta_salida}")
```

# Merge all files into a single dataset

```python
import pandas as pd
import os

# Input folder
carpeta_entrada = "data-science-searches/linksnormalizedheaderscleanwithdatabasesource"

# Output folder
carpeta_salida = "data-science-searches/linksnormalizedheaderscleanwithdatabasesourceall"
os.makedirs(carpeta_salida, exist_ok=True)

# List to store DataFrames
dfs = []

# Iterate through all CSV files
for archivo in os.listdir(carpeta_entrada):
    if archivo.endswith(".csv"):
        ruta = os.path.join(carpeta_entrada, archivo)
        df = pd.read_csv(ruta)
        dfs.append(df)

# Merge all DataFrames
df_total = pd.concat(dfs, ignore_index=True)

# Save final dataset
archivo_salida = os.path.join(carpeta_salida, "links_normalized_clean_withdatabasesource_all.csv")
df_total.to_csv(archivo_salida, index=False)

print(f"Merged file saved at: {archivo_salida}")
```
